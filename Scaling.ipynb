```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

data2 = {
    "age":    [20, 30, 40, 50, 25],
    "salary": [20000, 50000, 80000, 200000, 35000]
}

df2 = pd.DataFrame(data2)

# StandardScaler
ss = StandardScaler()
df2[["age_scaled", "salary_scaled"]] = ss.fit_transform(df2[["age", "salary"]])

print(df2)
```

```python
mm = MinMaxScaler()
df2[["age_mm", "salary_mm"]] = mm.fit_transform(df2[["age", "salary"]])
print(df2)
```
