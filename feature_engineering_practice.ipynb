```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    "name":       ["Alice", "Bob", "Carlos", "Diana", "Eve"],
    "education":  ["Bachelors", "Masters", "PhD", "High School", "Bachelors"],
    "city":       ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai"],
    "age":        [25, 30, 35, 28, 32],
    "salary":     [50000, 80000, 120000, 45000, 75000]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df.head)
```

```python
df.drop("name", axis=1)
```

```python
df = pd.DataFrame(data) # Re-initialize df to ensure 'education' is string
Oe = OrdinalEncoder(categories=[["High School", "Bachelors", "Masters", "PhD"]])
df["education"] = Oe.fit_transform(df[["education"]])
```

```python
df = df.drop(columns=["name"])
print(df.head())
```

```python
ohe = pd.get_dummies(df["city"], prefix = "city")
df = pd.concat([df,ohe], axis = 1)
df.drop(columns=["city"], inplace=True)
print(df)
```

```python
ss = StandardScaler()
df[["age", "salary"]] = ss.fit_transform(df[["age", "salary"]])
print(df)
```
