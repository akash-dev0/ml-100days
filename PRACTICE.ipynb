```python
import numpy as np
import pandas as pd
```

```python
data = {
    "hours_studied": [2, 4, 6, 8, 10, 3, 7, 5, 9, 1],
    "sleep_hours":   [6, 7, 8, 8, 9, 5, 7, 6, 8, 5],
    "attendance":    [60, 70, 80, 90, 95, 65, 85, 75, 92, 55],
    "grade":         ["C", "B", "B", "A", "A", "C", "A", "B", "A", "C"],
    "score":         [50, 65, 70, 85, 95, 55, 88, 72, 91, 45]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
#EDA
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull())
```

```python
#encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["grade"] = le.fit_transform(df["grade"])
print(df)
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df["hours_studied"] = ss.fit_transform(df[["hours_studied"]])
df["sleep_hours"] = ss.fit_transform(df[["sleep_hours"]])
df["attendance"] = ss.fit_transform(df[["attendance"]])
print(df)
```

```python
#Train test split
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = df.drop("score", axis = 1)
y = df["score"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Predict:", y_pred)
print("Actual:", y_test)
```

```python
#Metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, root_mean_squared_error
```

```python
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
MAE = mean_absolute_error(y_test, y_pred)
R2 = r2_score(y_test, y_pred)
```

```python
n = len(y_test)
k = x_test.shape[1]
adj_r2 = 1 - (1 - R2) * (n - 1)/(n - k - 1)
```

```python
print("MSE:", MSE)
print("RMSE:", RMSE)
print("MAE:", MAE)
print("R2:", R2)
print("Adjusted R2:", adj_r2)
```
