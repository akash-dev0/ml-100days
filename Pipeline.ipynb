```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
```

```python
data = {
    "age": [25, 30, np.nan, 40, 35, np.nan, 28, 45],
    "salary": [50000, 80000, 60000, np.nan, 70000, 90000, np.nan, 120000],
    "bought": [0, 1, 0, 1, 1, 1, 0, 1]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
x = df[["age", "salary"]]
y = df["bought"]
```

```python
x = df[["age", "salary"]]
y = df["bought"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Pipeline
pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler())
])

# Diagnostic: Check if x_train is defined before use
print(f"Attempting to process x_train. Is x_train defined? {'x_train' in globals()}")

x_train = pipe.fit_transform(x_train)
x_test = pipe.transform(x_test)

print("X_train processed :", x_train)
print("X_test processed :", x_test)
```
