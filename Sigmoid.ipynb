```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
y = np.array([1, 0, 1, 0])

w0 = np.zeros(2)
b  = 0
eta = 0.1

for i in range(len(X)):
    z     = np.dot(w0, X[i]) + b
    y_hat = sigmoid(z)
    error = y[i] - y_hat

    w0 = w0 + eta * error * X[i]
    b  = b  + eta * error

    print(f"i={i} | yi={y[i]} | y_hat={y_hat:.3f} | error={error:.3f} | w={w0}")
```
