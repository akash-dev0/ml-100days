```python
import numpy as np
```

```python
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
m = 0
b = 0
lr = 0.01
epochs = 1000
n = len(X)
```

```python
for i in range(epochs):
  y_pred = m*X + b

  slope_b = -2 * np.sum(y - y_pred)
  slope_m = -2 * np.sum((y - y_pred) * X)

  b = b - lr * slope_b
  m = m - lr * slope_m
  if i % 100 == 0:
        loss = np.sum((y - y_pred)**2)
        print(f"Epoch {i} → Loss: {loss:.4f}, m: {m:.4f}, b: {b:.4f}")
```

```python
print("m =", m)
print("b =", b)
```
