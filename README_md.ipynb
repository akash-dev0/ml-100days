```python
# 🏠 House Price Prediction

## Overview
A Machine Learning project that predicts house prices
using the Boston Housing dataset.

## What I did
- Performed EDA on 506 houses with 14 features
- Found strongest correlations with house price
- Built a Linear Regression model using top 2 features
- Achieved MAE of $3,890

## Key Findings from EDA
- More rooms (rm) = higher price (+0.7 correlation)
- More poverty (lstat) = lower price (-0.7 correlation)

## Tech Stack
- Python
- Pandas, NumPy
- Matplotlib, Seaborn
- Scikit-learn

## Model Results
- Algorithm: Linear Regression
- Features used: rm, lstat
- MAE: 3.89
```
