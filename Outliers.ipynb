```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "score": [45, 50, 48, 52, 47, 49, 51, 200, 46, 53]

}
```

```python
df = pd.DataFrame(data)
```

```python
#Visualize
sns.boxplot(df["score"])
plt.title("Score Distribution with Outlier")
plt.show()
```

```python
# IQR Method
Q1 = df["score"].quantile(0.25)
Q3 = df["score"].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower boundary: {lower}")
print(f"Upper boundary: {upper}")
```

```python
#REMOVE OUTLIER
df_clean = df[(df["score"] >= lower) & (df["score"] <= upper)]
print("Before:", len(df))
print("After:", len(df_clean))
```
