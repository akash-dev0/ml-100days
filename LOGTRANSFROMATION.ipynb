```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
salary = np.array([25000, 30000, 28000, 32000, 27000, 2500000])
```

```python
fig, axes = plt.subplots(1,2, figsize=(12, 4))
```

```python
#Before Trandsformation
axes[0].hist(salary, bins=10)
axes[0].set_title("Before Log Tranformation")
```

```python
#After Transformation
axes[1].hist(np.log(salary), bins=10)
axes[1].set_title("After Log Transform")
```

```python
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

salary = np.array([25000, 30000, 28000, 32000, 27000, 2500000])

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Before transformation
axes[0].hist(salary, bins=10)
axes[0].set_title("Before Log Transform")

# After transformation
axes[1].hist(np.log(salary), bins=10)
axes[1].set_title("After Log Transform")

plt.show()
```
