```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#our data
size = np.array([1000, 2000, 3000, 4000, 5000]).reshape(-1,1) #-1 means "figure it out yourself NumPy.-1 means "figure it out yourself NumPy."
                                                              #You're saying:
                                                              #I don't know how many rows — you calculate it
                                                              #I want 1 column
price = np.array([50, 90,  130, 170, 210])
```

```python
print(size)
print(price)
```

```python
#create and train the model
model = LinearRegression()
model.fit(size,price)
```

```python
#predict price for 2500 sqft
predicted = model.predict([[2500]])
print(f"Predicted price for 2500 sqft : {predicted[0]} lakhs")
```

```python
#print slope and intercept
print(f"Slope (m): {model.coef_}")
print(f"Intercept (c): {model.intercept_}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(size, price, color="blue", label="Actual data")
plt.plot(size, model.predict(size), color="red", label="Regression line")
plt.xlabel("Size (sqft)")
plt.ylabel("Price (lakhs)")
plt.title("House Price Prediction")
plt.legend()
plt.show()
```
